## Visualization & Explanatory Analysis

### Pertanyaan 1:

Hari Valintine dirayakan setiap tanggal 14 Februari

```python
df_jawaban=df_hour[(df_hour['dteday'] == '2011-02-14')]
print("Distribusi data padda Hari Valentine Tahun 2011")
df_plot = df_jawaban.groupby("hr")["cnt"].sum()
df_plot

plt.figure(figsize=(10, 6))
plt.plot(df_plot.index, df_plot.values, marker='o', linestyle='-')

plt.title('Distribusi Penyewaan Sepeda pada Hari Valentine 2011')
plt.xlabel('Jam (Hour)')
plt.ylabel('Total Penyewaan Sepeda')

plt.grid(True)
plt.show()
```

Pada jam 00:00-05:59 di hari Valentine, hanya ada 27 orang yang merental sepeda, dimana pada jam 02:00-04:59 merupakan titik terendah orang merental sepeda dimana setiap jam hanya ada 2 sepeda yang berhasil dirental. Ini dikarenakan, orang-orang masih tidur. Perentalan sepeda mulai naik cukup signifikan sampai jam 09:59, dimana puncaknya kenaikan sepeda terjadi pada jam 08:00-08:59 sebanyak 256. Ini dikarenakan banyak orang yang mulai bangun tidur dan memulai aktivitas mereka, seperti berangkat kerja, berangkat sekolah, berbelanja, dan lain sebagainya. Hari Valentine bukan merupakan hari libur bagi orang Portugal sehingga orang-orang menjalankan aktivitasnya seperti biasa. Pada pukul 10:00-11:59, terjadi penurunan rental sepeda. Pada saat jam makan siang, yaitu jam 12:00-13:59, perentalan sepeda meningkat. Pada jam 14:00-15:59, rental sepeda menurun tetapi tidak signifikan. Pada jam 16:00-18:00, rental sepeda pada Hari Valentine meningkat. Ini bisa terjadi karena jam ini adalah waktunya orang pulang kerja atau jam dimana orang pergi berkencan karena Hari Valentine. Pada jam 19:00-23:59, terjadi penurunan yang cukup stabil untuk perentalan sepeda.

### Pertanyaan 3:

```python
df_jawaban = df_day
df_jawaban["season"] = df_jawaban.season.apply(lambda x: "Spring" if x == 1 else ("Summer" if x==2 else ("Fall" if x==3 else "Winter")))
df_plot = df_jawaban.groupby(by='season').cnt.sum()
print(df_plot)

plt.figure(figsize=(10, 6))
df_plot.plot(kind='bar', color=["#90CAF9", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"])

plt.title('Total Penyewaan Sepeda berdasarkan Musim (Tahun 2012)')
plt.xlabel('Musim')
plt.ylabel('Total Penyewaan Sepeda')

plt.show()
```

Dari data yang sudah dianalisis, orang banyak merental sepeda pada musim gugur. Kita bisa memberikan beberapa asumsi mengapa musim gugur lebih banyak orang yang merental sepeda, yaitu karena suhu udara yang tidak begitu panas dan dingin serta orang lebih suka dengan vibe warna kecoklatan pada musim gugur. Pada musim semi, tidak banyak orang yang merental sepeda. Bisa diasumsikan bahwa musim semi pergantian dari musim dingin ke musim panas sehingga orang kurang suka bersepeda saat cuaca masih cukup dingin.

Cara meningkatkan jumlah rental sepeda pada musim semi:
- Mengadakan acara tur sepeda pada musim semi dengan hadiah yang menarik, seperti uang, iphone, tablet,dan lain-lain.
- Mengadakan promosi atau diskon khusus untuk penyewaan sepeda pada musim semi.
- Membuat dan menayangkan iklan tentang penyewaan sepeda, misal: iklan orang naik sepeda pada musim semi keliling kota.
- Menawarkan paket perentalan sepeda yang mencakup perlindungan terhadap cuaca, misal: mantel atau topi pelindung dari hujan ringan.

## Conclusion

- Distribusi rental sepeda pada Hari Valentine 2011 menunjukan puncak orang beraktivitas ada pada jam 08:00-08:59 sebanyak 256 dan orang jarang beraktivitas pada jam 02:00-04:59 dimana setiap jam hanya ada 2 sepeda yang berhasil dirental.
- Jumlah orang casual yang merental sepeda di musim kemarau pada hari kerja dan cuaca cerah pada tahun 2012 adalah 42.935 orang.
- Musim dimana orang paling banyak merental sepeda adalah musim gugur. Musim dimana orang paling sedikit merental sepeda adalah musim semi. Perentalan sepeda bisa ditingkatkan dengan adanya diskon, membuat iklan, mengadakan tur sepeda, dan menawarkan paket rental sepeda dengan mencakup pelindung terhadap cuaca.
